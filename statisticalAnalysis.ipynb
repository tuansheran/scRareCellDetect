```python
import scanpy as sc
from scipy.io import mmread
import numpy as np

data = mmread('scRNA.mtx')

sparse_matrix_coo = data.tocoo()


#variance, std and mean for gene experssion levels
mean_expression = np.mean(sparse_matrix_coo.data)
variance_expression = np.var(sparse_matrix_coo.data)
std_dev_expression = np.std(sparse_matrix_coo.data)

print(mean_expression)
print(variance_expression)
print(std_dev_expression)
```
